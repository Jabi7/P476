```python
n = 10000
s = [5, 10, 20, 100]

# rn.seed(n)
# x = [0]
# for i in range(N):
#     r = rn.random()
#     x.append(x[i]-1 if r < 0.5 else x[i]+1)
# x
```

```python
def rw(N = 5):
    x = [0]
    for i in range(N):
        r = rn.random()
        x.append(x[i]-1 if r < 0.5 else x[i]+1)
    return x[-1]
```

## Simulation

```python
for N in s:
    p = [0]*(2*N+1)
    x = [-N + i for i in range(2*N+1)]
    for i in range(n):
        r = rw(N)
        # print(N + r, r)
        p[N + r] += 1/n
        
    plt.bar(x,p)
    plt.title("N = "+str(N))
    plt.xlabel('x')
    plt.ylabel('P(x,N)')
    plt.show()
```

## From equation

```python
for N in s:

    x = [-N + i for i in range(2*N+1)]
    p = [f(N)/(2**N * f((2*N+i)) * f((2*N-i))) for i in x]
        
    plt.bar(x,p)
    plt.title("N = "+str(N))
    plt.xlabel('x')
    plt.ylabel('P(x,N)')
    plt.show()
```
